# Leitura dos dados - Camada Silver
Exploracao dos arquivos Parquet gerados na camada Silver.

```python
from pathlib import Path
import pandas as pd

SILVER_DIR = Path("../data/silver")
```

## Vendas (consolidado 2020-2022)

```python
vendas = pd.read_parquet(SILVER_DIR / "vendas.parquet")

print("Shape:", vendas.shape)
print("\nDtypes:")
print(vendas.dtypes)
vendas.head()
```

```python
vendas.describe()
```

```python
print("Nulos por coluna:")
vendas.isnull().sum()
```

```python
print("Periodo das vendas:")
print("De:", vendas["Data da Venda"].min())
print("Ate:", vendas["Data da Venda"].max())
```

## Devolucoes

```python
devolucoes = pd.read_parquet(SILVER_DIR / "devolucoes.parquet")

print("Shape:", devolucoes.shape)
print("\nDtypes:")
print(devolucoes.dtypes)
devolucoes.head()
```

```python
print("Motivos de devolucao:")
devolucoes["Motivo Devolução"].value_counts()
```

## Cadastro de Clientes

```python
clientes = pd.read_parquet(SILVER_DIR / "clientes.parquet")

print("Shape:", clientes.shape)
print("\nDtypes:")
print(clientes.dtypes)
clientes.head()
```

```python
print("Genero:")
print(clientes["Genero"].value_counts())
print("\nEstado Civil:")
print(clientes["Estado Civil"].value_counts())
print("\nNivel Escolar:")
print(clientes["Nivel Escolar"].value_counts())
```
