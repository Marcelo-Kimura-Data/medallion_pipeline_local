# Leitura dos dados - Camada Bronze
Exploracao inicial dos arquivos Parquet gerados na camada Bronze.

```python
from pathlib import Path
import pandas as pd

BRONZE_DIR = Path("../data/bronze")
```

## Vendas

```python
vendas_2020 = pd.read_parquet(BRONZE_DIR / "Base Vendas - 2020.parquet")
vendas_2021 = pd.read_parquet(BRONZE_DIR / "Base Vendas - 2021.parquet")
vendas_2022 = pd.read_parquet(BRONZE_DIR / "Base Vendas - 2022.parquet")

vendas = pd.concat([vendas_2020, vendas_2021, vendas_2022], ignore_index=True)

print("Shape:", vendas.shape)
print("\nDtypes:")
print(vendas.dtypes)
vendas.head()
```

```python
vendas.describe()
```

```python
print("Nulos por coluna:")
vendas.isnull().sum()
```

## Devolucoes

```python
devolucoes = pd.read_parquet(BRONZE_DIR / "Base Devoluções.parquet")

print("Shape:", devolucoes.shape)
print("\nDtypes:")
print(devolucoes.dtypes)
devolucoes.head()
```

```python
print("Motivos de devolucao:")
devolucoes["Motivo Devolução"].value_counts()
```
