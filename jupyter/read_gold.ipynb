# Leitura dos dados - Camada Gold
Exploracao das tabelas analiticas geradas na camada Gold.

```python
from pathlib import Path
import pandas as pd

GOLD_DIR = Path("../data/gold")
```

## Fato Vendas

```python
fato = pd.read_parquet(GOLD_DIR / "fato_vendas.parquet")

print("Shape:", fato.shape)
print("\nColunas:", list(fato.columns))
fato.head()
```

```python
print("Totais gerais:")
print("Valor total vendido: R$", round(float(fato["valor_total"].sum()), 2))
print("Custo total:         R$", round(float(fato["custo_total"].sum()), 2))
print("Lucro total:         R$", round(float(fato["lucro"].sum()), 2))
```

## Vendas por Periodo

```python
periodo = pd.read_parquet(GOLD_DIR / "vendas_por_periodo.parquet")

print("Shape:", periodo.shape)
periodo
```

```python
print("Melhor mes (lucro):")
print(periodo.loc[periodo["lucro"].idxmax()].to_string())
```

## Vendas por Loja

```python
lojas = pd.read_parquet(GOLD_DIR / "vendas_por_loja.parquet")

print("Shape:", lojas.shape)
lojas.head(10)
```

```python
print("Vendas por continente:")
lojas.groupby("Continente")[["valor_total", "lucro"]].sum().sort_values("lucro", ascending=False)
```

```python
print("Vendas por tipo de loja:")
lojas.groupby("Tipo")[["valor_total", "lucro", "qtd_vendida"]].sum().sort_values("lucro", ascending=False)
```

## Devolucoes por Motivo

```python
devolucoes = pd.read_parquet(GOLD_DIR / "devolucoes_por_motivo.parquet")

print("Shape:", devolucoes.shape)
devolucoes
```

```python
print("Devolucoes por motivo (total):")
devolucoes.groupby("Motivo Devolução")[["qtd_devolvida", "num_ocorrencias"]].sum().sort_values("qtd_devolvida", ascending=False)
```
